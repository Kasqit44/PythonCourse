```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
import seaborn as sns

j=1
cmap = ["pink", "wheat"]
for i in cmap:
    sns.distplot(np.random.pareto(j, 100), hist=False, color=i, 
                 kde_kws={"shade": True}, label="a="+str(j))
    sns.despine()
    j+=1
```
