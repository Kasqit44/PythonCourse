```python
import numpy as np
import pandas as pd
from sklearn import datasets
from matplotlib import pyplot as plt
from matplotlib import cm
import seaborn as sns
```

#### E1: Нарисовать функцию распределения Пуассона (np.random.poisson) 

```python
j=1
cmap = ["pink", "wheat", "r", "g", "b"]
for i in cmap:
    sns.distplot(np.random.poisson(j, 10000), hist=False, color=i, 
                 kde_kws={"shade": True}, label="lam="+str(j))
    sns.despine()
    j+=1
```

####  E2: Нарисовать функцию распределения Паретто (np.random.pareto)

```python
j=1
cmap = ["pink", "wheat"]
for i in cmap:
    sns.distplot(np.random.pareto(j, 100), hist=False, color=i, 
                 kde_kws={"shade": True}, label="a="+str(j))
    sns.despine()
    j+=1
```

#### E3: Построить графики зависимости цены жилья в Бостоне от различных параметров. 

```python
pd.set_option('display.max_rows', 10)

boston = datasets.load_boston()
boston_df = pd.DataFrame(boston['data'], columns=boston['feature_names'])
boston_df['PRICE'] = boston.target
boston_df
```

```python
for i in boston_df: 
    if "PRICE" not in i:
        sns.relplot(x="PRICE", y=i, data=boston_df)
        sns.despine()
```
