```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
import seaborn as sns

j=1
cmap = ["pink", "wheat", "r", "g", "b"]
for i in cmap:
    sns.distplot(np.random.poisson(j, 10000), hist=False, color=i, 
                 kde_kws={"shade": True}, label="lam="+str(j))
    sns.despine()
    j+=1
```
